## Задание 1
### 1.1 Создать новый признак price_per_sq_lot, который будет содержать среднюю стоимость за один кв. метр общей площади

```python
import pandas as pd
```

```python
df = pd.read_csv('kc_house_data.csv')
df.head(5)
```

```python
df['price_per_sq_lot'] = df['price']/df['sqft_lot']
df.head()
```

### 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома

```python
renovation = lambda x, y: y-x if y != 0 else 0
df['delta_renovated'] = df.apply(lambda x: renovation(x.yr_built, x.yr_renovated), axis=1)
df.head()
```

### 1.3 Создайте признаки года продажи, месяца продажи

```python
df['year_sold'] = df['date'].apply(lambda x: int(x[:4]))
df['month_sold'] = df['date'].apply(lambda x: int(x[4:6]))
df.head()
```

### 1.4 Удалите признаки date, zipcode, lat, long

```python
df.drop(columns=['date', 'zipcode', 'lat', 'long'], inplace=True)
```

## Задание 3
Составьте несколько сводных таблиц
### 3.1 Найдите среднюю стоимость домов в зависимости от количества спален. Отсортируйте от меньшей стоимости к большей

```python
price_bedrooms = df.groupby('bedrooms').aggregate('price').mean().reset_index().sort_values('price')
price_bedrooms
```

### 3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
price_state = df.groupby('condition').agg({'price': ['min', 'mean', 'max']})
price_state
```

### 3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
houses_view = df.groupby(['waterfront', 'view']).aggregate('id').count()
houses_view
```

```python
df.pivot_table(index='waterfront', columns='view',
               values='id', aggfunc='count',
               fill_value=0)
```

### 3.4 Каких домов в зависимости от этажности и количества спален больше?

```python
houses_floor_rooms = df.groupby(['floors', 'bedrooms']).aggregate('id').count().reset_index().sort_values('id', ascending=False)
houses_floor_rooms
```

```python
df.pivot_table(index='bedrooms', columns='floors',
               values='id', aggfunc='count',
               fill_value=0)
```

### 3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома

```python
houses_condition = df.groupby(['condition', 'grade']).aggregate('price').median()
houses_condition
```

```python
df.pivot_table(index='condition', columns='grade',
               values='price', aggfunc='median',
               fill_value=0)
```
